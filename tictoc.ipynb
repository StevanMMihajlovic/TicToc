```python

def display(r1,r2,r3):
    
    board = r1+r2+r3
    print("Current state: ")
    print("\n")
    print("\n")
    print('   |   |')
    print(' ' + board[0] + ' | ' + board[1] + ' | ' + board[2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[3] + ' | ' + board[4] + ' | ' + board[5])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[6] + ' | ' + board[7] + ' | ' + board[8])
    print('   |   |')
    print("\n")
```

```python

def user_choice(gamelist):
    
    x = input("Please enter where you want to put a new mark (1-9)?: ")
    while (not x.isdigit() or int(x)>9 or int(x)<1 or gamelist[int(x)-1]!= " "):
        if(not x.isdigit()):
            print("Insert integer!")
        elif (int(x)>9 or int(x)<1):
            print("Insert integer in specific range!")
        elif (gamelist[int(x)-1]!= " "):
            print("That field is taken!")          
        x = input("Please try another one! ")
    return int(x)
```

```python
def gameon_choice():
    
    choice = ""
    print("\n")
    while choice not in ["y", "Y", "n", "N"]:
        choice = input("Do you want to continue this game? (Y or N) ")
    print("\n")
    if choice == "y" or choice=="Y":
        return True
    else:
        print("GAME OVER")
        return False
```

```python
def replace_choice(gamelist, position, st):
    
    gamelist[position] = st
    return gamelist
```

```python
from IPython.display import clear_output

def tic_toc():
    
    turn=1
    game = True
    winner = False
    gamelist1=[" ", " ", " "]
    gamelist2=[" ", " ", " "]
    gamelist3=[" ", " ", " "]

    print("Welcome to Tik Toc game!")
    print("\n")
    p1 = input("Player 1, do you want to be X or O? ")
    while p1 not in ["X","x","O","o"]:
        p1 = input("Invalid input! Player 1, do you want to be X or O? ")
    print("\n")
    if p1=="x" or p1=="X":
        p2 = "O"
        p1 = "X"
        print("Player 1, you are X!")
        print("Player 2, you are O!")
    else: 
        p2 = "X"
        p1 = "O"
        print("Player 1, you are O!")
        print("Player 2, you are X!")
    
    print("\n")
    
    while game and not winner:
        
        
        if turn==1:
            print("Player 1, your move!")
            st = p1
        else:
            print("Player 2, your move!")
            st = p2

        display(gamelist1, gamelist2, gamelist3)
        position = user_choice(gamelist1+gamelist2+gamelist3)
        if position in range(1,4):
            gamelist1 = replace_choice(gamelist1, position-1, st)
        elif position in range(4,7):
            gamelist2 = replace_choice(gamelist2, position-3-1, st)
        else:
            gamelist3 = replace_choice(gamelist3, position-6-1, st)
        clear_output()
        display(gamelist1, gamelist2, gamelist3)
        winner = check_game(gamelist1,gamelist2,gamelist3, p1)
        if winner!= True:
            game = gameon_choice()
            clear_output()    
        if turn ==1:
            turn =2
        else:
            turn =1
```

```python
def check_game(r1,r2,r3,p1):
    
    draw = True
    for i in range(0,3):
        if (r1[i]==" " or r2[i]==" " or r3[i]==" "):
            draw = False
    if draw:
        print("GAME OVER, we have a DRAW!")
        return True
    if (r1[0]==r1[1] and r1[1]==r1[2] and r1[0]!=" "):
        if r1[0]==p1:
            print("GAME OVER, we have a WINNER: PLAYER 1!")
        else:
            print("GAME OVER, we have a WINNER: PLAYER 2!")
        return True
    if (r2[0]==r2[1] and r2[1]==r2[2] and r2[0]!=" "):
        if r2[0]==p1:
            print("GAME OVER, we have a WINNER: PLAYER 1!")
        else:
            print("GAME OVER, we have a WINNER: PLAYER 2!")
        return True
    if (r3[0]==r3[1] and r3[1]==r3[2] and r3[0]!=" "):
        if r3[0]==p1:
            print("GAME OVER, we have a WINNER: PLAYER 1!")
        else:
            print("GAME OVER, we have a WINNER: PLAYER 2!")
        return True
    for i in range(0,3):
        if (r1[i]==r2[i] and r2[i]==r3[i] and r1[i]!=" "):
            if r1[i]==p1:
                print("GAME OVER, we have a WINNER: PLAYER 1!")
            else:
                print("GAME OVER, we have a WINNER: PLAYER 2!")
            return True
    if (r1[0]==r2[1] and r2[1]==r3[2] and r1[0]!=" "):
        if r1[0]==p1:
            print("GAME OVER, we have a WINNER: PLAYER 1!")
        else:
            print("GAME OVER, we have a WINNER: PLAYER 2!")
        return True
    if (r1[2]==r2[1] and r2[1]==r3[0] and r1[2]!=" "):
        if r1[2]==p1:
            print("GAME OVER, we have a WINNER: PLAYER 1!")
        else:
            print("GAME OVER, we have a WINNER: PLAYER 2!")
        return True
    
```
